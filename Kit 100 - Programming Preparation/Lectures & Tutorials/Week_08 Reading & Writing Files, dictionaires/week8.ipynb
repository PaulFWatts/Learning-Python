### Read line by line using while

```python
freader = open("myfile.txt","r")
line = freader.readline()
while line !="":
    print(line)
    line = freader.readline()
freader.close()
```

There are special characters in the file that are not visible, for example "\n" is new line character that makes the spaces above.

```python
freader = open("myfile.txt","r")
line = freader.readline()
while line !="":
    for c in line:
        print(repr(c)," ",end="")
    print()
    line = freader.readline()
freader.close()
```

So, to print each line while ignoring the "\n" character, we use end="" in the print function.

```python
freader = open("myfile.txt","r")
line = freader.readline()
while line !="":
    print(line,end="")
    line = freader.readline()
freader.close()
```

### Read line by line using For

```python
freader = open("myfile.txt","r")
count = 1
for line in freader:
    print(count,":",line,end="")
    count +=1 # this is the same as count = count+1
    
freader.close()
```

### Move the reading position

```python
freader = open("myfile.txt","r")
freader.seek(200)
line = freader.readline()
print(line)

freader.close()
```

### Displaying file function

```python
def display_file(filename):
    freader = open(filename,"r")
    line = freader.readline()
    while line !="":
        print(line,end="")
        line = freader.readline()
    freader.close()
```

```python
display_file("myfile.txt")
```

### Modify content of a file

```python
freadwrite = open("newfile.txt","r")
content = freadwrite.read()
freadwrite.close()
freadwrite = open("newfile.txt","w")
newcontent = content.replace("I want to be a full-stack developer","I want to be a front-end developer")
freadwrite.write(newcontent)
freadwrite.flush()
freadwrite.close()
```

```python
display_file("newfile.txt")
```

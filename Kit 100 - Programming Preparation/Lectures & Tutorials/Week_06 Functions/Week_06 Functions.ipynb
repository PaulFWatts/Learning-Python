## 6. Functions

1. Order of functions is not important but must be defined before it is called.
2. Can be a void function or a value returning function
3. Use `:` to end the function definition
4. Function statements must be indented
5. Can define a main function and call it (see slides)
6. Can have compulsorily parameters and optional parameters (see slides) 
7. Can return multiple values, use “_” if don’t require a specific return variable
8. Can declare global variables inside a function
